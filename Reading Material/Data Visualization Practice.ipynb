```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
fine_name = [i for i in os.listdir() if i.endswith('csv')][0]
```

```python
data = pd.read_csv(fine_name)
```

```python
display(data.head(10))
```

```python
plt.scatter(data['day'], data['tip'], c= data['size'], s = data['total_bill'])
plt.title('Scatter Plot')
plt.xlabel('Day of Week')
plt.ylabel('Amount of Tip')
plt.colorbar()
plt.show()
```

```python
plt.plot(data["tip"])
plt.plot(data["size"])
plt.xlabel('Day')
plt.ylabel('Tip')
plt.show()
```

```python
plt.bar(data['day'], data["tip"])
plt.xlabel('Day')
plt.ylabel('Tip')
plt.show()
```

```python
data[['tip', 'day']]

plt.plot(data.groupby('day')['tip'].max())
```

```python
plt.hist(data['total_bill'])
plt.title('Histogram of Bill amount')
plt.xlabel('Bill Amount')
plt.ylabel('No. of Customers')
```
